```python
def y (x):
    sin = math.sin(x)+1
    return math.sqrt(sin)
y(6)
```

```python
def y (x):
    z = math.sin(x) / math.cos(x)
    #برای اینکه این تابع به ازای اعداد منفی ارور میده از شرط استفاده میکنیم
    if z < 0:
        return "یه عدد دیگه وارد کن"
    return math.sqrt(z)

y(1)
```

```python
def y (x):
    op = math.tan(x) / math.sin(x)
        #برای اینکه این تابع به ازای اعداد منفی ارور میده از شرط استفاده میکنیم
    if op < 0:
        return "یه عدد دیگه وارد کن"
    return math.sqrt(op)

y(5)
```

```python
def y (x):
    op = 1/1+math.exp(-x)
    return math.sqrt(op)

y(7)
```

```python
def y (a,w):
    sqrtW = math.sqrt(w)
    sqrtA = math.sqrt(a)
    return w-a*sqrtW / sqrtA

y(4,9)
```
